```python
import sys
import pandas as pd
import matplotlib
import numpy as np
import scipy as sp
import IPython
import sklearn
import mglearn
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import Ridge
X,y = mglearn.datasets.load_extended_boston()
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)

ridge = Ridge().fit(X_train,y_train)
print("training set score: {:.2f}".format(ridge.score(X_train,y_train)))
print("test set score: {:.2f}".format(ridge.score(X_test,y_test)))
```

```python
ridge10 = Ridge(alpha = 0.1).fit(X_train,y_train)
print("training set score: {:.2f}".format(ridge10.score(X_train,y_train)))
print("test set score: {:.2f}".format(ridge10.score(X_test,y_test)))
```

```python
ridge01 = Ridge(alpha = 0.2).fit(X_train,y_train)
print("training set score: {:.2f}".format(ridge01.score(X_train,y_train)))
print("test set score: {:.2f}".format(ridge01.score(X_test,y_test)))
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

plt.plot(ridge10.coef_, '^', label="Ridge alpha=0.1")
#plt.plot(ridge.coef_, 's', label="Ridge alpha=1")
plt.plot(ridge01.coef_, 'v', label="Ridge alpha=0.2")
#plt.plot(lr.coef_, 'o', label="LinearRegression")  # Changed to lr
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")  # Corrected method name
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])  # Corrected variable name
plt.xlim(xlims)
plt.ylim(-25, 25)
plt.legend()
plt.show()  # Added to display the plot
```
