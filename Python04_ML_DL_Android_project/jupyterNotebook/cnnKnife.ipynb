```python
import os
import cv2

# 저장할 폴더 생성
os.makedirs('resized/knifes', exist_ok=True)

knifes = os.listdir('C:/Users/Jounghyoun/Downloads/Knife_vs_Pistol/Knife/')
i = 0

for knife in knifes:
    sample = cv2.imread(f'C:/Users/Jounghyoun/Downloads/Knife_vs_Pistol/Knife/{knife}')
    if sample is not None:
        resized = cv2.resize(sample, (227, 227))
        gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
        cv2.imwrite(f'resized/knifes/{i}.jpg', gray)
    i += 1
```

```python
import random as rd
import cv2
import matplotlib.pyplot as plt

# 5개의 이미지를 한 줄에 표시하는 함수 정의
def display_images_in_row(images, titles):
    fig, axes = plt.subplots(1, 5, figsize=(15, 3))
    for i, (img, title) in enumerate(zip(images, titles)):
        axes[i].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axes[i].set_title(title)
        axes[i].axis('off')
    plt.show()

# 5개의 이미지씩 여러 줄에 표시
for _ in range(5):
    images = []
    titles = []
    for _ in range(5):
        n = rd.randint(1, 320)
        k_img = cv2.imread(f'resized/knifes/{n}.jpg')
        images.append(k_img)
        titles.append(f'Shape: {k_img.shape}')
    display_images_in_row(images, titles)
    print('\n')
```
